```python
import cv2
import numpy as np

# Load the image
img = cv2.imread('3.png', 0)

# Apply a blur to the image to reduce noise
blur = cv2.GaussianBlur(img, (5,5), 0)

# Apply the Hough Circle Transform
circles = cv2.HoughCircles(blur, cv2.HOUGH_GRADIENT, 1, 20, param1=50, param2=30, minRadius=2, maxRadius=30)

# Check if any circles were detected
if circles is not None:
    # Convert the (x, y) coordinates and radius of the circles to integers
    circles = np.round(circles[0, :]).astype("int")
    
    # Initialize counters for total and partial circles
    total_circles = 0
    partial_circles = 0
    
    # Loop through the detected circles
    for (x, y, r) in circles:
        # Draw the circle on the image
        cv2.circle(img, (x, y), r, (0, 255, 0), 2)
        
        # Check if the circle is complete or partial
        mask = np.zeros_like(img)
        cv2.circle(mask, (x, y), r, 255, -1)
        masked_img = cv2.bitwise_and(img, mask)
        non_zero_pixels = cv2.countNonZero(masked_img)
        if non_zero_pixels == 0:
            partial_circles += 1
        else:
            total_circles += 1
    
    # Display the image with detected circles
    cv2.imshow('Detected Circles', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    # Print the total number of circles and partial circles detected
    print('Total Circles:', total_circles)
    print('Partial Circles:', partial_circles)
else:
    print('No circles were detected in the image.')
```

```python
import cv2
import numpy as np
import torch
import detectron2
from detectron2.config import get_cfg
from detectron2.engine import DefaultPredictor
from detectron2.modeling import build_model
from detectron2.checkpoint import DetectionCheckpointer

# Load the image
img = cv2.imread('3.png')

# Initialize the Detectron2 configuration
cfg = get_cfg()
cfg.merge_from_file('detectron2/configs/COCO-Detection/faster_rcnn_R_101_FPN_3x.yaml')
cfg.MODEL.WEIGHTS = 'detectron2://COCO-Detection/faster_rcnn_R_101_FPN_3x/138204752/model_final_298dad.pkl'
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5

# Build and load the model
model = build_model(cfg)
model.eval()
checkpointer = DetectionCheckpointer(model)
checkpointer.load(cfg.MODEL.WEIGHTS)

# Initialize the predictor
predictor = DefaultPredictor(cfg)

# Run the predictor on the image
outputs = predictor(img)

# Extract the circles from the output
circles = outputs['instances'].pred_boxes.tensor.cpu().numpy()
radii = outputs['instances'].pred_classes.cpu().numpy()

# Check if any circles were detected
if len(circles) > 0:
    # Initialize counters for total and partial circles
    total_circles = 0
    partial_circles = 0
    
    # Loop through the detected circles
    for i in range(len(circles)):
        # Get the center coordinates and radius of the circle
        x = int((circles[i][0] + circles[i][2]) / 2)
        y = int((circles[i][1] + circles[i][3]) / 2)
        r = int(radii[i])
        
        # Draw the circle on the image
        cv2.circle(img, (x, y), r, (0, 255, 0), 2)
        
        # Check if the circle is complete or partial
        mask = np.zeros_like(img)
        cv2.circle(mask, (x, y), r, 255, -1)
        masked_img = cv2.bitwise_and(img, mask)
        non_zero_pixels = cv2.countNonZero(masked_img)
        if non_zero_pixels == 0:
            partial_circles += 1
        else:
            total_circles += 1
    
    # Display the image with detected circles
    cv2.imshow('Detected Circles', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    # Print the total number of circles and partial circles detected
    print('Total Circles:', total_circles)
    print('Partial Circles:', partial_circles)
else:
    print('No circles were detected in the image.')
```

```python
import torch
import cv2
import numpy as np
from detectron2.config import get_cfg
from detectron2.engine import DefaultPredictor
from detectron2.model_zoo import model_zoo

# Load the image
img = cv2.imread('image.jpg')

# Set up the configuration and predictor for CircleNet
cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_101_FPN_3x.yaml"))
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-Detection/faster_rcnn_R_101_FPN_3x.yaml")
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.7
cfg.MODEL.CIRCLE_ON = True
predictor = DefaultPredictor(cfg)

# Use CircleNet to detect circles in the image
outputs = predictor(img)
circles = outputs["instances"].to("cpu").get("pred_circles").numpy()

# Check if any circles were detected
if circles is not None:
    # Initialize counters for total and partial circles
    total_circles = 0
    partial_circles = 0
    
    # Loop through the detected circles
    for (x, y, r) in circles:
        # Draw the circle on the image
        cv2.circle(img, (int(x), int(y)), int(r), (0, 255, 0), 2)
        
        # Check if the circle is complete or partial
        mask = np.zeros_like(img)
        cv2.circle(mask, (int(x), int(y)), int(r), 255, -1)
        masked_img = cv2.bitwise_and(img, mask)
        non_zero_pixels = cv2.countNonZero(masked_img)
        if non_zero_pixels == 0:
            partial_circles += 1
        else:
            total_circles += 1
    
    # Display the image with detected circles
    cv2.imshow('Detected Circles', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    # Print the total number of circles and partial circles detected
    print('Total Circles:', total_circles)
    print('Partial Circles:', partial_circles)
else:
    print('No circles were detected in the image.')
```

```python
import cv2
import numpy as np

# Load the image
img = cv2.imread('Sample.png')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to the image to reduce noise
gray = cv2.GaussianBlur(gray, (5, 5), 0)

# Apply the HoughCircles function to detect circles in the image
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 1, 20, param1=100, param2=30, minRadius=2, maxRadius=30)

# Check if any circles were detected
if circles is not None:
    # Convert the (x, y) coordinates and radius of the circles to integers
    circles = np.round(circles[0, :]).astype("int")
    
    # Initialize a counter for the number of circles detected
    num_circles = 0
    
    # Loop through each detected circle
    for (x, y, r) in circles:
        # Draw the circle on the image
        cv2.circle(img, (x, y), r, (0, 255, 0), 2)
        
        # Increment the counter for the number of circles detected
        num_circles += 1
        
        # Print the dimensions of the circle
        print('Circle', num_circles, ': x =', x, 'y =', y, 'r =', r)
        
    # Display the image with the detected circles
    cv2.imshow("Detected Circles", img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    # Print the total number of circles detected
    print('Total Circles:', num_circles)
else:
    print('No circles were detected in the image.')
```
